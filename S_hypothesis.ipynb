```python
# 데이터 표준화 #

import numpy as np

a=[0,3,5,7,10]

from scipy import stats
stats.zscore(a)
```

```python
import pandas
aaa = pandas.read_csv('cars.csv')
aaa.head()
```

```python
# 표준정균분포 확률 #

# 누적 확률 #

y1=stats.norm.cdf(-1)
y1

y2=stats.norm.cdf(1.0) - stats.norm.cdf(-1.0)
y2

print ('y1=', y1)
print ('y3=', y2)

# 역누적 확률 #

y3 = stats.norm.ppf(0.9)
y3
print ('y3=',y3)
```

```python
# 정규 분포 그래프 #

import matplotlib.pyplot as plt

rv1 = stats.norm(0,1)                  # 평균=0, 표준편차 =1 #
x1 = np.linspace(-5,5,100)
y4 = rv1.pdf(x1)

plt.plot(x1,y4)
plt.ylim(0,0.5)
plt.show()
```

```python
# 회사 연봉 평균 1000, 표준편차 60 #

rv2 = stats.norm(1000,60)                  # 평균 = 1000, 표준편차 = 60 #
x2 = np.linspace(700,1300,1000)
y5 = rv2.pdf(x2)

plt.plot(x2,y5, color='r')
plt.ylim(0,0.01)
plt.show()

rv_950 = rv2.cdf(950)
print ('950이하 확률=',round(rv_950*100,3))
print ('950이상 확률=',round((1-rv_950)*100,3))

rv_90 = rv2.ppf(0.9)
print ('누적 90% 연봉값 =',round(rv_90,2))
```
